```python
import pandas as pd

filename = 'mutations.csv'
data = pd.read_csv(filename)

# print out the table
data.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# y is the target variable (aka the patients)
# grabs the first column and checks if it starts with 'C' and assigns 1 if true, 0 if false
y = data['Unnamed: 0']
y = y.apply(lambda x: 1 if x.startswith('C') else 0)
print(y)

# drops the first column
# results in just the mutation data 
X = data.drop(columns=['Unnamed: 0'])
print(X)

# Step 2: Split the data into training and test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,     # 20% for testing
    train_size=0.8,    # 80% for training
    random_state=42,   # Ensures reproducibility
    shuffle=True,      # Shuffle the data before splitting
    #stratify=y,        # Keep the same proportion of classes
)

# Step 3: Create a RandomForestClassifier and fit it to the training data
# defines the classifier and the options i put on it
clf = RandomForestClassifier(
    n_estimators=300,       # Number of trees
    max_depth=30,           # Maximum depth of trees
    random_state=42,        # Reproducible results
)
#actually builds the forest of trees using the training data
clf.fit(X_train, y_train)

# Step 4: Select one of the trees (e.g., the first tree)
tree = clf.estimators_[0]

#Step 5: Visualize the selected tree
plt.figure(figsize=(20, 10))
plot_tree(tree, 
          filled=True, 
          feature_names=X.columns,  # Use the feature column names
          class_names=['Non-Cancer', 'Cancer'],  # Use class names (0, 1)
          rounded=True)
plt.show()

# Step 6: Test the classifier on the test set
y_pred = clf.predict(X_test)

# Step 7: Calculate accuracy (how many times it gets it right)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Feature importances
importances = clf.feature_importances_

# Print the top 10 important features
important_features = sorted(zip(importances, X.columns), reverse=True)[:10]
print("Top 10 important features:")
for importance, feature in important_features:
    print(f"{feature}: {importance:.4f}")
```
