```python
# Author : Kara Nemet
# Date : 9/29/2024
# Purpose: Read in a csv file and anaylze the data using pandas to answer given questions about the data
# shoutout to kaleb demaline and kaleb keichel for helping me learn pandas/python !!

import pandas as pd

data = pd.read_csv('mutations.csv', index_col=0)

data 
```

```python
# data.shape[1] gives the number of columns in the data frame

print("Number of Unique Mutations : " , data.shape[1])
```

```python
# data.shape[0] gives the number of rows in the data frame

print("Number of Unique Samples : ", data.shape[0])
```

```python
# data.loc["C1"] finds the row with the name "C1"
# .sum() adds up all the values in the row, giving us the number of mutations in that sample

print("Number of Mutations in Sample C1 : ", data.loc["C1"].sum())
```

```python
# data.loc["NC1"] finds the row with the name "NC1"
# .sum() adds up all the values in the row, giving us the number of mutations in that sample

print("Number of Mutations in Sample NC1 : ", data.loc["NC1"].sum())
```

```python
# data.sum().sum() adds up all the values in the data frame (whole table), giving us the total number of mutations
# data.shape[0] gives the number of rows in the data frame, giving us the number of samples
# total number of mutations / number of samples = average number of mutations per sample

print("Average Number of Mutations per Sample : ", data.sum().sum() / data.shape[0])
```

```python
# data.sum(axis=1) adds up all the values in each row, giving us the number of mutations in each sample
#  PSA : axis=1 represents columns
# .min() finds the minimum value in the list of number of mutations in each sample

print("Minimum number of mutations in a sample : ", data.sum(axis=1).min())
```

```python
# data.sum(axis=1) adds up all the values in each row, giving us the number of mutations in each sample
#  PSA : axis=1 represents columns
# .max() finds the maximum value in the list of number of mutations in each sample

print("Maximum number of mutations in a sample : ", data.sum(axis=1).max())
```

```python
# data.filter(like='BRAF', axis=1) finds all the columns that contain the string 'BRAF'
# .sum(axis=1) adds up all the values in each column, giving us the number of mutations in each sample containing the 'BRAF' gene
# .sum() adds up all the values in the list of number of mutations in each sample containing the 'BRAF' gene
# so basically, find "BRAF" genes --> add up mutations in each sample containing "BRAF" gene --> add up all mutations in all samples containing "BRAF" gene

print("Number of individuals who have a mutation in the ‘BRAF’ gene : ", data.filter(like='BRAF', axis=1).sum(axis=1).sum())
```

```python
# data.filter(like='KRAS', axis=1) finds all the columns that contain the string 'KRAS'
# .sum(axis=1) adds up all the values in each column, giving us the number of mutations in each sample containing the 'KRAS' gene
# .sum() adds up all the values in the list of number of mutations in each sample containing the 'KRAS' gene
# so basically, find "KRAS" genes --> add up mutations in each sample containing "KRAS" gene --> add up all mutations in all samples containing "KRAS" gene

print("Number of individuals who have a mutation in the 'KRAS' gene : ", data.filter(like='KRAS', axis=1).sum(axis=1).sum())
```

```python
# data.sum().sum() adds up all the values in the data frame (whole table), giving us the total number of mutations
# data.shape[1] gives the number of columns in the data frame, giving us the number of mutations
# total number of mutations / number of mutations = average number of individuals per mutation

print("Average number of individuals per mutation : ", data.sum().sum() / data.shape[1])
```

```python
# data.sum() gives the sum of each sample (row)
# the .min() function then gives the minimum value of the sum out of all the samples

print("Minimum number of individuals per mutation : ", data.sum().min())
```

```python
# data.sum() gives the sum of each sample (row)
# the .max() function then gives the maximum value of the sum out of all the samples

print("Maximum number of individuals per mutation : ", data.sum().max())
```
