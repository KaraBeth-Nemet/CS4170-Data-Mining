```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.2f}'.format

filename = 'mutations.csv'
data = pd.read_csv(filename, index_col=0)
samples = data.shape[0]
data.head()
```

# Calc TP, FP, TN, FN for ALL Mutations

```python
mutation_data = []

for mutation in data:
    curr_tp, curr_fp, curr_tn, curr_fn = 0, 0, 0, 0
    for index in data.index:
        if data[mutation][index] == 1:
            if index.startswith('C'):
                curr_tp += 1
            else:
                curr_fp += 1
        else: 
            if index.startswith('C'):
                curr_fn += 1
            else:
                curr_tn += 1
    mutation_data.append([mutation, curr_tp, curr_fp, curr_tn, curr_fn])

mutation_data 

tp_fp_tn_fn_chart = pd.DataFrame(columns=['Mutation', 'TP', 'FP', 'TN', 'FN'])
tp_fp_tn_fn_chart['Mutation']= [x[0] for x in mutation_data]
tp_fp_tn_fn_chart['TP']= [x[1] for x in mutation_data]
tp_fp_tn_fn_chart['FP']= [x[2] for x in mutation_data]
tp_fp_tn_fn_chart['TN']= [x[3] for x in mutation_data]
tp_fp_tn_fn_chart['FN']= [x[4] for x in mutation_data]

tp_fp_tn_fn_chart.head(10)
```

# Table of the top 10 features ranked by accuracy (with corresponding accuracy values)


```python
# Accuracy = (TP + TN) / (TP + FP + TN + FN)

tp_fp_tn_fn_chart['Accuracy (%)'] = ((tp_fp_tn_fn_chart['TP'] + tp_fp_tn_fn_chart['TN']) / (tp_fp_tn_fn_chart['TP'] + tp_fp_tn_fn_chart['FP'] + tp_fp_tn_fn_chart['TN'] + tp_fp_tn_fn_chart['FN'])) *100
tp_fp_tn_fn_chart.sort_values(by='Accuracy (%)', ascending=False).head(10)
```

# Find Most Useful Feature by ranking TP-FP

```python
mutation_data = []

for mutation in data:
    curr_tp, curr_fp = 0, 0
    for index in data.index:
        if data[mutation][index] == 1:
            if index.startswith('C'):
                curr_tp += 1
            else:
                curr_fp += 1
    mutation_data.append([mutation, curr_tp, curr_fp])

mutation_data.sort(key=lambda x: x[1] - x[2], reverse=True)
most_useful_mutation = mutation_data[0][0]
print(f"The most useful mutation is: {most_useful_mutation}")
```

# Find Most Useful Feature in Group A and B by ranking TP-FP

```python
# Get Group A and Group B patients
group_a_patients = []
group_b_patients = []

for sample in data.index:
    if data[most_useful_mutation][sample] == 1:
        group_a_patients.append(sample)
    else:
        group_b_patients.append(sample)

max_len = max(len(group_a_patients), len(group_b_patients))
for i in range(max_len - len(group_a_patients)):
    group_a_patients.append('')
for i in range(max_len - len(group_b_patients)):
    group_b_patients.append('')
```

```python
# Get Group A
group_a_mutation_data = []

# drop '' from group_a_patients
group_a_patients = [x for x in group_a_patients if x != '']
group_a_data = data.loc[group_a_patients]
for mutation in group_a_data:
    curr_tp, curr_fp = 0, 0
    for index in group_a_data.index:
        if group_a_data[mutation][index] == 1:
            if index.startswith('C'):
                curr_tp += 1
            else:
                curr_fp += 1
    group_a_mutation_data.append([mutation, curr_tp, curr_fp])

group_a_mutation_data.sort(key=lambda x: x[1] - x[2], reverse=True)

# drop most_useful_mutation from group_a_mutation_data to avoid duplicates
group_a_mutation_data = [x for x in group_a_mutation_data if x[0] != most_useful_mutation]

# Find the most useful mutation in Group A
group_a_most_useful_mutation = group_a_mutation_data[0][0]
print(f"The most useful mutation for group A is: {group_a_most_useful_mutation}")
```

```python
# Get Group B
group_b_mutation_data = []
# find the true positive and false positive for group A

#go into data and pull data for group A patients
# drop '' from group_a_patients
group_b_patients = [x for x in group_b_patients if x != '']

group_b_data = data.loc[group_b_patients]

for mutation in group_b_data:
    curr_tp, curr_fp = 0, 0
    for index in group_b_data.index:
        if group_b_data[mutation][index] == 1:
            if index.startswith('C'):
                curr_tp += 1
            else:
                curr_fp += 1
    group_b_mutation_data.append([mutation, curr_tp, curr_fp])


group_b_mutation_data.sort(key=lambda x: x[1] - x[2], reverse=True)

#drop the most useful mutation from the list
group_b_mutation_data = [x for x in group_b_mutation_data if x[0] != most_useful_mutation]


# Find the most useful mutation in Group B
group_b_most_useful_mutation = group_b_mutation_data[0][0]
print(f"The most useful mutation for group B is: {group_b_most_useful_mutation}")
```

# Result of classifying all samples

```python
# Decision Tree Classifier

# keep track of how many we get correct
total_tp, total_fp, total_tn, total_fn = 0, 0, 0, 0

for sample in data.index:
    if data.loc[sample, most_useful_mutation] == 1:
        if data.loc[sample, group_a_most_useful_mutation] == 1:
            print(f"Our tree says {sample} is classified as C")
            if sample.startswith('C'):
                total_tp += 1
            else :
                total_fp += 1
        else:
            print(f"Our tree says {sample} is classified as NC")
            if sample.startswith('NC'):
                total_tn += 1   
            else:
                total_fn += 1  
    else:
        if data.loc[sample, group_b_most_useful_mutation] == 1:
            print(f"Our tree says {sample} is classified as C")
            if sample.startswith('C'):
                total_tp += 1
            else: 
                total_fp += 1
        else:
            print(f"Our tree says {sample} is classified as NC")
            if sample.startswith('NC'):
                total_tn += 1
            else:
                total_fn += 1

```

# (including a confusion matrix)

```python
# make a confusion matrix with the results

print(f"Total TP: {total_tp}")
print(f"Total FP: {total_fp}")
print(f"Total TN: {total_tn}")
print(f"Total FN: {total_fn}")

confusion_matrix = pd.DataFrame({'Predicted Healthy': [total_tp, total_fp], ' Predicted Cancer': [total_fn, total_tn]}, index=['Actual Healthy', 'Actual Cancer'])
confusion_matrix

```

# Values of the metrics listed above 

```python
# accuracy is how close a given set of measurements (observations or readings) are to their true value
accuracy = (total_tp + total_tn) / (total_tp + total_fp + total_tn + total_fn) * 100

# sensitivity measures how well a test can identify true positives
sensitivity = total_tp / (total_tp + total_fn) * 100

# specificity measures how well a test can identify true negatives
specificity = total_tn / (total_fp + total_tn) * 100

# percision is how close the measurements are to each other
percision = total_tp / (total_tp + total_fp) * 100

# miss rate = rate at which a system or model makes false negatives
miss_rate = total_fn / (total_fn + total_tp) * 100

# false discovery rate = percentage of times a test incorrectly identifies a "discovery" or significant result.
fdr = total_fp / (total_fp + total_tp) * 100

# false omission rate tells you how often the model mistakenly classified something as negative when it should have been positive
forr = total_fn / (total_fn + total_tn) * 100

final_table = pd.DataFrame(columns=['Metric', 'Value'])

final_table['Metric'] = ['Accuracy (%)', 'Sensitivity (%)', 'Specificity (%)', 'Percision (%)', 'Miss Rate (%)', 'FDR (%)', 'FOR (%)']

final_table['Value'] = [accuracy, sensitivity, specificity, percision, miss_rate, fdr, forr]

final_table
```
